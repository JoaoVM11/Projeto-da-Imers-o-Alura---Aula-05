```python
df = pd.DataFrame(documents)
df.columns = ["Titulo", "Conteudo"]
df
```

```python
model = "models/embedding-001"
```

```python
def embed_fn(title, text):
  return genai.embed_content(model=model,
                                 content=text,
                                 title=title,
                                 task_type="RETRIEVAL_DOCUMENT")["embedding"]
```

```python
df["Embeddings"] = df.apply(lambda row: embed_fn(row["Titulo"], row["Conteudo"]), axis=1)
df
```

```python
def gerar_e_buscar_consulta(consulta, base, model):
  embedding_da_consulta = genai.embed_content(model=model,
                                 content=consulta,
                                 task_type="RETRIEVAL_QUERY")["embedding"]

  produtos_escalares = np.dot(np.stack(df["Embeddings"]), embedding_da_consulta)

  indice = np.argmax(produtos_escalares)
  return df.iloc[indice]["Conteudo"]
```

```python
consulta = "Qual é o valor do Diagnóstico Comercial?"

trecho = gerar_e_buscar_consulta(consulta, df, model)
print(trecho)
```

```python
generation_config = {
  "temperature": 0,
  "candidate_count": 1
}
```

```python
#Mesclando a resposta vinda pelo documento Embeddings com a IA Generativa para melhorar a resposta de acordo com o perfil do cliente
prompt = f"Reescreva essa informação de uma forma que, antes de informar o preço, incentive o cliente a adquirir o Diagnóstico Comercial. Lembrando que esse cliente terá um perfil corporativo, sendo um CEO, Diretor Comercial, Gerente, etc. Não esqueça de informar o preço e a duração do projeto no final. {trecho}"

model_2 = genai.GenerativeModel("gemini-1.0-pro",
                                generation_config=generation_config)
response = model_2.generate_content(prompt)
print(response.text)
```
